```python
from bs4 import BeautifulSoup
import requests
```

```python
web = "https://es.wikipedia.org/wiki/Copa_Am%C3%A9rica_2021"
respuesta = requests.get(web)
contenido = respuesta.text
```

```python
soup = BeautifulSoup(contenido, "lxml")
```

```python
matches = soup.find_all("table", class_="collapsible autocollapse vevent plainlist mw-collapsible mw-made-collapsible mw-collapsed")
```

```python
for match in matches:
  print(match.find("td", {"align": "right"}).get_text())
  print(match.find("td", {"align": "center"}).get_text())
  print(match.find("td", {"align": "left"}).get_text())



```
